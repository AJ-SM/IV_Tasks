```python
import gym
import numpy  as np
import time
import matplotlib.pyplot as plt 
import random
env = gym.make('MiniGrid-Empty-6x6-v0')
```

```python

#initilazation
no_of_step = 165
no_of_Episodes = 150
epsilon = 1.0
gamma = 0.98
alpha = 0.3
n_action = 3
destn =(4,4)
Q_sa= {}
for x in range(1,6):
    for y in range(1,6):
        for d in range(4):
            Q_sa[((x,y),d)] = {a:0 for a in range(n_action)}
```

```python
# Requried Function :
def TakeAction(env,action):
    obs,reward,done,_,_ = env.step(action)
    next_state = (env.agent_pos,obs['direction'])
    return reward,next_state,done

def eps_greedy(epsilon,state):
    if np.random.rand()<epsilon:
        #Explor
        
        action = np.random.randint(0,n_action)
    else:
        # Go Greedy 
        action = max(Q_sa[state],key=Q_sa[state].get)
    return action
```

```python
# SARSA Learning
rewards =[]
sts= []

for episode in range(no_of_Episodes):
    env.reset()
    epsilon = max(0.1, epsilon * 0.99)
    Gt=0
    steps = 0
    state = ((1, 1), 0) 
    action = eps_greedy(epsilon,state)
    print(action)
    
    
    for steps in range(no_of_step):
        reward,next_state,done = TakeAction(env,action)
        next_action = eps_greedy(epsilon,next_state)
        Q_sa[state][action] += alpha * (reward + gamma * Q_sa[next_state][next_action] - Q_sa[state][action])
        print(reward)
        state = next_state
        action = next_action
        Gt+=reward
        steps +=1
        if done:
            break
    rewards.append(Gt)
    sts.append(steps)
    
    
        
        
        

```
